```python
import json
import re
import csv
```

```python
input_file_path = '/content/How_I met_your_mother_episodes_dialogues.json'
output_file_path = '/content/How_I met_your_mother_episodes_dialogues_parsed.json'

# Функция для замены одинарных кавычек на двойные, используя регулярное выражение
def replace_quotes(match):
    # Если найденная строка начинается и заканчивается одинарной кавычкой, заменяем ее на двойную
    if match.group(1) == "'" and match.group(3) == "'":
        return '"' + match.group(2) + '"'
    # Иначе возвращаем исходную строку без изменений
    return match.group(0)

# Читаем содержимое файла
with open(input_file_path, 'r') as input_file:
    # Читаем содержимое файла
    json_content = input_file.read()

# Заменяем одинарные кавычки на двойные, используя регулярное выражение
json_content = re.sub(r"'([^']*)'", replace_quotes, json_content)

# Пытаемся распарсить содержимое файла как JSON
try:
    parsed_json = json.loads(json_content)
    # Если успешно, сохраняем в новый файл
    with open(output_file_path, 'w') as output_file:
        json.dump(parsed_json, output_file, indent=4)
    print("Файл успешно преобразован и сохранен в", output_file_path)
except json.JSONDecodeError as e:
    print("Ошибка при преобразовании файла:", e)
```

```python
with open('/content/drive/MyDrive/МФТИ/nlp2/chatbot/Data/How_I met_your_mother_episodes_dialogues_parsed.json', 'r') as file:
    data = json.load(file)

# Создать CSV файл и записать заголовки столбцов
with open('dialogues.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Q', 'A', 'Context']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    for episode in data:
        dialogues = episode['dialogues']
        for i in range(1, len(dialogues)):
            if dialogues[i]['name'] == 'Barney':
                # Первая реплика перед "Barney"
                Q = dialogues[i - 1]['text']
                # Реплика "Barney"
                A = dialogues[i]['text']
                # Реплика после "Barney" (если это возможно)
                next_index = i + 1
                next_dialogue = dialogues[next_index]['text'] if next_index < len(dialogues) else ""
                # Контекст: реплика перед "Barney" и реплика после него
                Context = f"{Q} | {next_dialogue}"
                # Записать в CSV файл
                writer.writerow({'Q': Q, 'A': A, 'Context': Context})
```
